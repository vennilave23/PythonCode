```python
import os
current_dir = os.getcwd()
print(current_dir)
```

```python
new_directory = os.path.join(current_dir, 'PythonClass')
if not os.path.exists(new_directory):
    os.makedirs(new_directory)
    print(f"Directory '{new_directory}' created successfully.")
else:
    print(f"Directory '{new_directory}' already exists.")

os.chdir(new_directory)
print(f"Current working directory changed to: {os.getcwd()}")
os.chmod(new_directory, 0o755)  # Setting permissions to read, write, execute for owner and read, execute for group and others
```

```python
os.chdir(os.path.expanduser('~') )
print(f"Current working directory changed to: {os.getcwd()}")
```

```python
os.chdir(new_directory)  # Change back to the original directory
print(f"Current working directory changed back to: {os.getcwd()}")
os.chdir("..")
print(f"Current working directory changed to parent directory: {os.getcwd()}")
```

```python
directory = os.getcwd()
file_name ="example.txt"
file_path = os.path.join(directory, file_name)
print(file_path)
content = "This is an example file created using Python."
with open(file_path, 'w') as file:
    file.write(content)
print(f"File '{file_name}' created at: {file_path}")

with open(file_path, 'r') as file:
    file_content = file.read()
print(f"Content of '{file_name}':\n{file_content}")

additional_content = "\nThis line is added to the existing content."    
with open(file_path, 'a') as file:
    file.write(additional_content)
print(f"Additional content appended to '{file_name}'.")

with open(file_path, 'r') as file:
    updated_content = file.read()
print(f"Updated content of '{file_name}':\n{updated_content}")
```

```python
def read_file(file_path):
  try:
    with open(file_path, 'r') as file:
        content = file.read()
    return content
  except FileNotFoundError:
    return f"Error: The file '{file_path}' was not found."
  except IOError:
    return f"Error: An I/O error occurred while reading the file '{file_path}'."    



def create_file(file_path, content):
    try:
        with open(file_path, 'w') as file:
            file.write(content)
        return f"File '{file_path}' created successfully."
    except IOError:
        return f"Error: An I/O error occurred while creating the file '{file_path}'."
    

def append_to_file(file_path, additional_content):
        try:
            with open(file_path, 'a') as file:
                file.write(additional_content)
            return f"Content appended to '{file_path}' successfully."
        except IOError:
            return f"Error: An I/O error occurred while appending to the file '{file_path}'." 
        

    

file_content = read_file(file_path)
print(file_content)

new_file_path = os.path.join(os.getcwd(), "create_example.txt") 
create_content = "This file is created using the create_file function."
result = create_file(new_file_path, create_content)
print(result) 
print(read_file(new_file_path))
additional_content = "\nThis content is appended using the append_to_file function."
append_result = append_to_file(new_file_path, additional_content) 
print(append_result)
print(read_file(new_file_path))


```
