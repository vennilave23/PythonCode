```python
# #A local fruit seller faces significant challenges in managing the billing process for customers. Currently, the seller uses a manual method, which is time-consuming, prone to errors, and lacks efficiency. This can lead to customer dissatisfaction and potential revenue loss due to mismanagement or billing inaccuracies.
# #To address this, the fruit seller is exploring the implementation of a basic Shopping Cart Management System as an initial test to improve the shopping experience. The system does not need to be overly complex but should provide essential functionalities to streamline operations and improve customer experience.


# Problem Breakdown - Steps involved for creating the Shopping Cart Management for the Fruit Seller

# Given Data : Inventory containing tuples with product ID, name, and price per kg of fruits. Then, display the inventory in a decorative format using headers and separators, with each item's details (ID, name, price) printed in a structured table format

# Shopping Cart Management Methods:

# add_items_to_cart: Write a Python function add_item_to_cart(item_id, Inventory) that adds an item from the Inventory (using item_id as index) to a shopping cart Your_Cart, and prints a message confirming the item has been added and the cart updated

# remove_item: Write a Python function remove_item(cart_index, cart_items) that removes an item from the cart_items list at the given cart_index and prints a message confirming the item has been removed and the cart updated

# total_cost: Write a Python function total_cost(cart_items) that calculates the total cost of all items in the cart_items list (each item being a tuple where the price is the third element) and prints the total cost in a formatted message.

# display_cart: Write a Python function display_cart(Your_Cart) that checks if the Your_Cart list is not empty. If the cart contains items, it prints a formatted list of the items (with their index, name, quanity, and price per kg). If the cart is empty, it prints a message saying the cart is empty.

# Main Function: (Centralize and navigate through all functionalities.) Write a Python program with a loop for a shopping cart system. Display the cart and prompt the user to: add items (add_item_to_cart), calculate total cost (total_cost), remove items (remove_item), or exit (option 9). Include input handling for invalid choices with appropriate messages.
```

```python
Inventory  = [
    (1, "Apples", 250),
    (2, "Cherry", 650),
    (3, "Chickoo", 50),
    (4, "Grapes", 90),
    (5, "Mangoes", 150),
]

# Decorative formats of displaying the Given Inventory
print("+" * 50)
print(f"{'Fruit Sellers Inventory'.center(50)}")
print("+" * 50)
print(f"{'Product ID':<15}{'Name':<20}{'Price Per KG':<10}")
print("-" * 50)

for item in Inventory:
     print(f"{item[0]:<15} {item[1]:<20} ${item[2]:<10.2f}")
print("-" * 50)

# Initailizing Your_Cart here
Your_Cart = []
```

```python
def add_item_to_cart(item_id, Inventory):
    '''
    Add the Items_id to your cart and display the message.
    '''
    try:
        item = Inventory[item_id - 1]
        Your_Cart.append(item)
        print(f"{'>'*10} The item '{item[1]}' has been added to the cart, and your shopping cart has been updated. {'<'*10} \n\n")
    except IndexError:
        print(f"Invalid item ID: {item_id}")

        ## Add Test Case

# Test Cases
# Test 1: Add a valid item
print("Test 1: Adding a valid item")
add_item_to_cart(1, Inventory)
print(f"Expected: 1 item in cart, Found: {len(Your_Cart)} item(s) in cart.\n")

# Test 2: Add another valid item
print("Test 2: Adding another valid item")
add_item_to_cart(2, Inventory)
print(f"Expected: 2 items in cart, Found: {len(Your_Cart)} item(s) in cart.\n")

# Test 3: Try to add an invalid item
print("Test 3: Adding an invalid item")
add_item_to_cart(99, Inventory)
print(f"Expected: Error message and 2 items in cart, Found: {len(Your_Cart)} item(s) in cart.\n")

```

```python
def remove_item(cart_index, cart_items):
    '''
    Remove item with given item_id and display the message.
    '''
    try:
        removed_item = cart_items.pop(cart_index - 1)
        print(f"{'>'*10} The item '{removed_item[1]}' has been removed from the cart, and your shopping cart has been updated. {'<'*10} \n\n")
    except IndexError:
        print("Invalid cart index or cart is empty.")
```

```python
# Test Cases
# Test 1: Remove a valid item
print("Test 1: Removing a valid item (index 2)")
remove_item(2, Your_Cart)
print(f"Expected: 1 items left in cart, Found: {len(Your_Cart)} item(s) in cart.\n")

# Test 2: Remove another valid item
print("Test 2: Removing another valid item (index 1)")
remove_item(1, Your_Cart)
print(f"Expected: 0 item left in cart, Found: {len(Your_Cart)} item(s) in cart.\n")

# Test 3: Try to remove an item with an invalid index
print("Test 3: Removing an item with an invalid index (index 10)")
remove_item(10, Your_Cart)
print(f"Expected: Error message and 0 item left in cart, Found: {len(Your_Cart)} item(s) in cart.\n")
```

```python
def total_cost(cart_items):
    '''
    Calculate the total cost of the items in the shopping cart and display the total value.
    '''
    total = 0
    for item in cart_items:
        total += item[2]
        # item_price = item[2]
        # total = total + item_price

    print("~"*40)
    print(f"Total cost : {total}".center(40) )
    print("~"*40, "\n\n")
    return total
```

```python
# Test Cases
# Test 1: Calculate total cost for a non-empty cart
Your_Cart = [(1, "Apples", 250), (2, "Cherry", 650), (3, "Chickoo", 50)]
print("Test 1: Non-empty cart")
total = total_cost(Your_Cart)
print(f"Expected: 950, Found: {total}\n")

# Test 2: Calculate total cost for an empty cart
Your_Cart = []
print("Test 2: Empty cart")
total = total_cost(Your_Cart)
print(f"Expected: 0, Found: {total}\n")

# Test 3: Calculate total cost for a cart with duplicate items
Your_Cart = [(1, "Apples", 250), (1, "Apples", 250), (3, "Chickoo", 50)]
print("Test 3: Cart with duplicate items")
total = total_cost(Your_Cart)
print(f"Expected: 550, Found: {total}\n")
```

```python
def display_cart(Your_Cart):
    # Displaying your cart items in decorative format if the cart is not empty
    if Your_Cart:
        print("+" * 50)
        print(f"{'Your Cart'.center(40)}")
        print("+" * 50)
        print(f"{'Index':<10} {'Name':<20} {'Quantity':<10} {'Price Per KG':<10}")
        print("-" * 50)

        unique_items = set(Your_Cart)  # Get unique items in the cart
        print(unique_items)
        total_cost = 0
        for index, item in enumerate(unique_items):
            quantity = Your_Cart.count(item)  # Calculate quantity of each item
            price_per_kg = item[2]
            total_price = price_per_kg * quantity
            print(f"{index+1:<10} {item[1]:<20} {quantity:<10} ${total_price:<10.2f}")
            total_cost += total_price
        print("-" * 50)
        print(f"{'Total Cost':<40} ${total_cost:<10.2f}")  # Display grand total

    else:
        print(f"{'>'*10} Your cart is empty !!!, Why not buy Fruits from us ... {'<'*10} \n\n")
```

```python
# Test 3: Display a cart with duplicate items
Your_Cart = [(1, "Apples", 250), (1, "Apples", 250), (2, "Cherry", 650)]
print("Test 3: Cart with duplicate items")
display_cart(Your_Cart)
print("\n")
```

```python
def main():
    while True:
        print("\n" + "="*50)
        print("Welcome to the Fruit Seller's Shopping Cart System".center(50))
        print("="*50)
        display_cart(Your_Cart)
        print("\nPlease choose an option:")
        print("1. Add item to cart")
        print("2. Remove item from cart")
        print("3. Calculate total cost")
        print("9. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            try:
                item_id = int(input("Enter the Product ID to add to cart: "))
                add_item_to_cart(item_id, Inventory)
            except ValueError:
                print("Invalid input. Please enter a valid Product ID.")
        elif choice == '2':
            try:
                cart_index = int(input("Enter the index of the item to remove from cart: "))
                remove_item(cart_index, Your_Cart)
            except ValueError:
                print("Invalid input. Please enter a valid index.")
        elif choice == '3':
            total_cost(Your_Cart)
        elif choice == '9':
            print("Thank you for shopping with us! Goodbye!")
            break
        else:
            print("Invalid choice. Please select a valid option.")


```

```python
if __name__ == "__main__":
    main()
```
