```python
inventory_data = [ 
 (1, "Apple", 50),
 (2, "Banana", 100),
 (3, "Orange", 75) ,
 (4, "Grapes", 200), 
(5, "Mango", 150),
]

print("+", "-" * 50, "+")
print(f"{'Fruit Sellers Inventory'.center(50)}")
print("+", "-" * 50, "+")
print("| {:<15} | {:<20} | {:10}".format("ID", "Product Name", "Quantity"))
print("+", "-" * 50, "+")
for item in inventory_data:
    print("| {:<15} | {:<20} | {:10}".format(item[0], item[1], item[2]))
print("+", "-" * 50, "+")   


def add_new_product(inventory, product_id, product_name, quantity):
    inventory.append((product_id, product_name, quantity))
    print(f"Added new product: {product_name} with ID {product_id} and quantity {quantity}")

def update_product_quantity(inventory, product_id, new_quantity):
    for index, item in enumerate(inventory):
        if item[0] == product_id:
            inventory[index] = (item[0], item[1], new_quantity)
            print(f"Updated product ID {product_id} with new quantity {new_quantity}")
            return
    print(f"Product ID {product_id} not found in inventory.")

def display_inventory(inventory):
    print("+", "-" * 50, "+")
    print(f"{'Fruit Sellers Inventory'.center(50)}")
    print("+", "-" * 50, "+")
    print("| {:<15} | {:<20} | {:10}".format("ID", "Product Name", "Quantity"))
    print("+", "-" * 50, "+")
    for item in inventory:
        print("| {:<15} | {:<20} | {:10}".format(item[0], item[1], item[2]))
    print("+", "-" * 50, "+")

def remove_product(inventory, product_id):
    for index, item in enumerate(inventory):
        if item[0] == product_id:
            removed_item = inventory.pop(index)
            print(f"Removed product: {removed_item[1]} with ID {product_id}")
            return
    print(f"Product ID {product_id} not found in inventory.")

def add_items_to_cart(cart, inventory, product_id, quantity):
    for item in inventory:
        if item[0] == product_id:
            if item[2] >= quantity:
                cart.append((product_id, item[1], quantity))
                print(f"Added {quantity} of {item[1]} to cart.")
                return
            else:
                print(f"Not enough stock for {item[1]}. Available quantity: {item[2]}")
                return
    print(f"Product ID {product_id} not found in inventory.")

def view_cart(cart):
    print("+", "-" * 50, "+")
    print(f"{'Shopping Cart'.center(50)}")
    print("+", "-" * 50, "+")
    print("| {:<15} | {:<20} | {:10}".format("ID", "Product Name", "Quantity"))
    print("+", "-" * 50, "+")
    for item in cart:
        print("| {:<15} | {:<20} | {:10}".format(item[0], item[1], item[2]))
    print("+", "-" * 50, "+")

def remove_items_from_cart(cart, product_id):
    for index, item in enumerate(cart):
        if item[0] == product_id:
            removed_item = cart.pop(index)
            print(f"Removed {removed_item[2]} of {removed_item[1]} from cart.")
            return
    print(f"Product ID {product_id} not found in cart.")

    

```
